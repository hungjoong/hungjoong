##### 02 정렬 : sort_values
- df.sort_values([열1, 열2, ..., 열n], ascending=bool)
1. 1개의 열을 기준으로 오름차순 정렬
1. 1개의 열을 기준으로 내림차순 정렬 ascending = False
1. 2개 이상의 열을 기준으로 정렬
1. 2개 이상의 열기준 오름&내림 섞어 정렬 ascending = 리스트형식

```python
import pandas as pd

a1 = [60, 84, 80, 70, 19]
a2 = [77, 62, 95, 85, 17]
a3 = [61, 97, 72, 67, 15]
a4 = [75, 65, 95, 51, 18]
cols = ["국어", "영어", "수학", "과학", "나이"]
df = pd.DataFrame([a1, a2, a3, a4], index=list('ABCD'), columns=cols)
df
```

```python
df.sort_values(["수학", "나이"], ascending=[0, 1])
```

- df.mask(조건문, 참일때의 값)
- df.sort_values([열1, 열2, ..., 열n], ascending=bool)

```python
import pandas as pd

df1 = pd.read_excel("E10EXAMPLE.xlsx")
df1["정렬"] = df1["나이"].mask(df1["나이"]<60, 0)
# '정렬' 컬럼을 만들고 '나이' 컬럼의 값을 넣어준다. 그리고 나이가 60 이하이면 '0'을 넣어준다.
df1 = df1.sort_values(["점수", "정렬", "나이"], ascending=[0,0,1])
df1.drop("정렬", axis=1).to_clipboard(index=False)
```

##### 06 pivot & 피벗테이블
1. df를 피벗테이블로 만들어 본다 pivot_table
1. df를 멀티인덱스인 피벗테이블로 만든다 pivot_table
1. 숫자가 아닌 문자가 담긴 열을 value로 df를 피벗해본다 pivot

- df.povot_table(열, index=[열1,...,열n], columns=[열1,...,열n], aggfunc=집계함수) 
- df.povot(values=열, index=[열1,...,열n], columns=[열1,...,열n]) 

```python
import pandas as pd
a1 = ['송중기', '이유빈', '권보아', '차유람', '박보검', '김범수',
      '이효리', '유승호', '김유정']
a2 = list("남여여여남남여남여")
a3 = list("AAABBBCCC")
a4 = [97, 88, 78, 64, 84, 89, 87, 92, 99]
a5 = [i + '등' for i in list("123321321")]
a6 = [i + '시' for i in list("수정정수수수정정수")]
a7 = ['이름', '성별', '반', '점수', '반등수', '비고']

df = pd.DataFrame([a1, a2, a3, a4, a5, a6], index=a7, columns=list(range(1, 10))).T
df["점수"] = df["점수"].astype(int)
df
```

```python
import pandas as pd
df1 = pd.read_excel("E14EXAMPLE.xlsx")
df2 = pd.read_excel("E14EXAMPLE.xlsx", sheet_name=2)
a1 = ["1위", "2위", "3위", "4위"]

#stack으로 언피벗

pdf1 = df1.loc[:, "1위":"4위"].stack().reset_index()
pdf1.columns = ["횟수", "순위", "국가"]
pdf1 = pdf1.pivot_table("횟수", index="국가", columns="순위", aggfunc="count")
pdf1.sort_values(a1, ascending=False).iloc[:8].to_clipboard()


#melt로 언피벗

pdf2 = df1.melt(id_vars=["개최국"], value_vars=a1, var_name="순위", value_name="국가")
pdf2 = pdf2.loc[pdf2["개최국"].isin(df2["국가"])]
pdf2 = pdf2.pivot_table("개최국", index="국가", columns="순위", aggfunc="count")
pdf2.sort_values(a1, ascending=False).iloc[:8].to_clipboard()
```

```python
import pandas as pd
df1 = pd.read_excel("E15EXAMPLE.xlsx")
df2 = pd.read_excel("E15EXAMPLE.xlsx", sheet_name=1)
df3 = pd.read_excel("E15EXAMPLE.xlsx", sheet_name=2)

#작년명부에서 연차를 전부 1씩 더해주기
df1["연차"] = df1["연차"] + 1

#은퇴명부(df3)에 속한 은퇴선수데이터를 df1에서 삭제
cond1 = df1["이름"].isin(df3["이름"])
df1 = df1.loc[~cond1]

#신인지명일람(df2)에서 롯데선수만 필터링
cond2 = df2["지명팀"] == "롯데"
df2 = df2.loc[cond2]

# df1과 df2를 df1의 형식으로 합치기
df3 = pd.concat([df1, df2.reindex(columns=df1.columns, fill_value=1)])
df3.to_clipboard(index=False)
```

##### 특강03 groupby로 동명이인 구분하기

```python
import pandas as pd
# 변수지정코드
a1 = ['2008-07-23', '송중기', 88]
a2 = ['2008-09-21', '김나현', 72]
a3 = ['2008-05-06', '박효신', 78]
a4 = ['2008-03-05', '송중기', 93]
a5 = ['2008-08-13', '김나현', 91]
a6 = ['2008-12-24', '송중기', 85]
df = pd.DataFrame([a1, a2, a3, a4, a5, a6], columns=["생일", "이름", "점수"])
df["생일"] = pd.to_datetime(df["생일"])

# 동명이인끼리 생일순으로 rank 매기기
df["rank"] = df.groupby("이름")["생일"].rank()

# 매긴 랭크결과를 문자열로 바꾸기
df["rank문자"] = df["rank"].astype("int").astype("str")

# 동명이인 존재여부를 True, False로, keep=False 로 동명이인을 전부 True로
df["dup"] = df.duplicated("이름", keep=False)

# 불리언마스킹으로  dup열이 True이면 동명이인넘버링을 하고 아니면 이름을 반환
df["이름2"] = df["이름"].mask(df["dup"], df["이름"] + df["rank문자"])

# 결과에서 최초 df가 가지고있던 생일, 이름, 점수열만 인덱싱
df2 = df[["생일", "이름2", "점수"]]

# 이름2열의 이름을 이름열로 바꾸기
df2.columns = ["생일", "이름", "점수"]
```

```python
import pandas as pd
# 변수지정코드 생략

# 동명이인끼리 생일순으로 rank 매겨서 그걸 문자로 만든 시리즈를 만든다
s1 = df.groupby("이름")["생일"].rank().astype("int").astype("str")

# duplicated로 동명이인 존재여부를 조건문으로 만들어 동명이인 있을때만 이름열에 s1 값을 추가로 마스킹한다
df["이름"] = df["이름"].mask(df.duplicated("이름", keep=False), df["이름"] + s1)
```
