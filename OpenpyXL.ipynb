```python
#io와 pandas 모듈 import
import io
import pandas as pd
```

```python
# 드라이브 파일 경로 변수
Namefile = '직원명단.xlsx'
Sourcefile = 'source.xlsx'
```

```python
#pandas read_csv로 불러오기
data = pd.read_excel(Namefile)
data.head()
```

```python
# load_workbook 모듈을 사용한다
from openpyxl import load_workbook

# 표지마다 넣을 직원명단 엑셀을 불러온다. 워크북(wb)으로 먼저 불러오고 시트명을 지정해서 워크시트(ws) 변수에 담아준다.
wb = load_workbook(Namefile)
ws = wb['Sheet1']

# 표지 양식인 소스(source.xlsx)를 불러온다. 그리고 표지마다 직원명단과 인덱스번호가 들어간 최종(dest) 파일로 넣어준다.
new_wb = load_workbook(Sourcefile)
dest = new_wb['Sheet1']

# 자 이제 for loop 수행하면서 지금은 예제로 10명만 했지만, 100명, 1000명 그 이상 순식간에 표지를 만들어 낸다.
for i in range(2, 12):
  dest['F1'] = ws.cell(i, 1).value
  dest['C23'] = ws.cell(i, 3).value
  new_wb.copy_worksheet(dest)

# 최종(dest) 파일을 잘 정리해주면, 오늘 업무 끝.
new_wb.save('dest.xlsx')
new_wb.close()

# 오늘도 수고하셨습니다. 퇴근합시다.
```
